```python
#functionssssssssssssssssssssssssssssss

import cmath
import numpy as np
import pandas as pd
from bokeh.io import show, output_notebook, output_file
from bokeh.plotting import figure
output_notebook()

# Th_1 = 0.7
# Th_2 = 0.5

# #-------------------------------------------------------------self
def cluster_self(cluster_cluster):
    for ind_c,c in enumerate(cluster_cluster.CC):
        for ind_ci,ci in enumerate(c):
            if ind_c != ind_ci and OR(cluster_cluster.Point[ind_c],cluster_cluster.Point[ind_ci])!=0:
            #print ind_c,ind_ci
            #print 'OR',OR(data.Point[ind_c],data.Point[ind_ci])
            #print 'AND',AND(data.Point[ind_c],data.Point[ind_ci])
                cluster_cluster.CC[ind_c][ind_ci] = round(float(AND(cluster_cluster.Point[ind_c]
                                                        ,cluster_cluster.Point[ind_ci]))/float(OR(cluster_cluster.Point[ind_c]
                                                        ,cluster_cluster.Point[ind_ci])), 3)
        
#print cluster_cluster
    cluster_cluster = cluster_cluster.sort_values(by='Select',ascending=False)
#print cluster_cluster.Select.head(50)

#-------------------------------------------------------------self


#--------------------------------------------------------------------------self
def OR(a,b):
    OR = 0
    for i in range(len(a)):
        if (a[i] or b[i]):
            OR += 1
    return OR
def AND(a,b):
    AND = 0
    for i in range(len(a)):
        if (a[i] and b[i]):
            AND += 1
    return AND

def S_merge(data):
    for ind_c,c in enumerate(data.CC):
        ii = 0
        for ci in c:
            if ci > 0:
                ii+=1
        data.loc[ind_c, 'self_W'] = sum(c)
        data.loc[ind_c, 'self_L'] = float(ii)
        data.loc[ind_c, 'Select'] = round(data.self_W[ind_c]/float(ii),3)
    s1 = max(data.self_L)
    s2 = max(data.self_W)
    for ind_c in range(len(data)):
        #data.loc[ind_c, 'self_L'] = round((data.self_L[ind_c]/s1),3)
        #data.loc[ind_c, 'self_W'] = round((data.self_W[ind_c]/s2),3)
        data.loc[ind_c, 'Select_n'] = round((data.self_W[ind_c]/s2)/(float(data.self_L[ind_c]/s1)),3)
    return data

#--------------------------------------------------------------------------------------self


#R-------------------------------------------------------------------------
def R(N,a,b,c):
    CM = c/cmath.sqrt((a*b))
    R = (N*CM-cmath.sqrt(a*b))/(cmath.sqrt((N-a)*(N-b)))
    return R.real
#CR_0---------------------------------------------------------------
def CR_0():
    #Rm = []
    for index0,c in enumerate(C):
        r = []
        for index1,cc in enumerate(C):
            if data.Count[index0] != 0 and data.Count[index1] != 0:
                cover = 0
                N1 = 0
                N2 = 0
                for j in range(Data):
                    cover += c[j] & cc[j]
                    N1 += c[j]
                    N2 += cc[j] 
                RR = R(Data,N1,N2,cover)
                if RR > 0:
                    r.append(round(RR,3))
                else:
                    r.append(round(RR,3))
            else:
                r.append(-1)
            #print 'r',index0,index1,r
            data.set_value(index0,'R' , r)

    return 0

#MAX-----------------------------------------------------------------
def T(last):

    TEMP = []
    for t in data.R:
        for index,i in enumerate(t):
            if i == last:
                t[index] = 0
        TEMP.append(sorted(set(t),reverse=True))

    find_m = []
    for t in TEMP:
        if t != []:
            find_m.append(max(t))
    #print sorted(find_m,reverse=True)
    MAX = max(find_m)
    if MAX < Th_1:                                                                     #*********************Th_1
        MAX = 0
        #print 'MMM'

    return MAX
#Select_Merge---------------------------------------------------------------------------------------
def Select_Merge(data,last,cluster_cluster):
    CR_0()
    #cluster_self(cluster_cluster)

    MAX = T(last)
    print 'merge score:',MAX
    
    if MAX == 0:
        data = data.drop('R', axis=1)
        return data

    merge_cand = []
    for index,rt in enumerate(data.R):
        for i,r in enumerate(rt):
            if r == MAX:
                #print index,i
                if sorted([index,i]) not in merge_cand:
                    merge_cand.append(sorted([index,i]))

    #print '\n'
    #print MAX
    #print 'merge_cand',merge_cand
    #print '\n'

#Merge First Pair=============================
    if len(merge_cand)!= 0:
        
        chose = 0
        if merge_cand[0][0] in cluster_cluster.Id.head(50) and merge_cand[0][1] in cluster_cluster.Id.head(50):
            chose = 0
            print 'YES!'
        print "~",chose
        

        i = merge_cand[chose][0]
        j = merge_cand[chose][1]
        print 'M>',merge_cand[chose]
        if data.Count[i] == 0:
            i = data.IdTo[i]
            #print 'IdToi',i
        if data.Count[j] == 0:
            j = data.IdTo[j]
            #print 'IdToj',j
        
        for k in range(len(data.Merge[j])):
            #print data.Merge[j][k]
            data.Merge[i].append(data.Merge[j][k])

        data.loc[i,'Count' ]+= len(data.Merge[j])
        data.set_value(j,'Merge', data.Merge[i])
        data.loc[j,'Count' ] = 0
        data.loc[j,'IdTo'] = i
        data.set_value(i,'Point', [sum(x) for x in zip(data.Point[i], data.Point[j])] )
        
    #print data

    return data

#@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@-----------select

#Choice_Double-------------------------------------------------------
def Choice_Double(double):
    s=-1
    for u in double:
        if u > 1:
            s = u
            break

    #s = min(double)
    #M = max(double)
    if s != -1:
        choice = double.index(s)
        #print double[choice]
        double[choice] = 0
        return choice
    else:
        return -1
#Select where to GO----------------------------------------------------------------
def double_select(data,selected_cluster,choice,RESULT,mode):
    Mm = []
    Mm1 = []#compare which to go
    Mm2 = []
    Mm3 = []
    Mm4 = []
    rank = []
    rank1 = []#final to keep in rank[0]
    rank2 = []
    rank3 = []
    rank4 = []
    
    record = []
    print 'c-------------->',choice
    if choice == -1:
        return RESULT
    
    for index,n in enumerate(selected_cluster):
        a=0
        b=0
        if (data.Point[n][choice] != 0.0) :#data.Point
            print 'c: ',n,RESULT[n][choice],data.Point[n][choice]
            record.append(n)
            a = data.Count[n]
            b = sum(data.Point[n])

            A = (b-data.Point[n][choice])/(a-1) # sum(data.Point[n]) #without
            B = b/a  #data.Point[n][choice]                             #add
            C = sum(data.Point[n])
            D = data.Point[n][choice]
            
            #print '---',B-A,D/C,(B-A)*(-D/C),'---'
        else:
            A = 1
            B = -100
            C = 1
            D = -100
       
        if mode == 1:
            CQ = B-A
            if CQ >= 0:
                print 'CQ!!!=>',CQ
        elif mode == 2:
            CQ = D/C
        elif mode == 3:
            CQ = (B-A)*(-D/C)
        elif mode == 4:
            CQ = (1/(B-A))*(-D/C)
        elif mode == 5:
            CQ = FUCK*(1/(B-A))+(1-FUCK)*(D/C)
            
        Mm.append(CQ)
        Mm1.append(B-A)
        Mm2.append(D/C)
        Mm3.append((B-A)*(-D/C))
        Mm4.append( FUCK*(1/(B-A))+(1-FUCK)*(D/C) )
        
        CQ = 0
        
        #B-A  D/C  (B-A)*(D/C)################################################################################---------MODE
        
        
    print 'Mm:',Mm
    print 'Mm1:',Mm1
    print 'Mm2:',Mm2
    print 'Mm3:',Mm3
    print 'Mm4:',Mm4
    
    
    re1 = max(Mm) #keep
    rank.append(Mm.index(re1))
    #print 're1',re1
    
    re11 = max(Mm1) #keep
    rank1.append(Mm1.index(re11))
    #print 're11',re11
    
    re2 = max(Mm2) #keep
    rank2.append(Mm2.index(re2))
    #print 're2',re2
    re3 = max(Mm3) #keep
    rank3.append(Mm3.index(re3))
    #print 're3',re3
    re4 = max(Mm4) #keep
    rank4.append(Mm4.index(re4))
    #print 're4',re4
    
    
    for i in range(len(Mm)):
        if Mm[i] != re1 and i in record:
            rank.append(selected_cluster[i])

    print '1:keep in',rank1[0]
    print '2:keep in',rank2[0]
    print '3:keep in',rank3[0]
    print '4:keep in',rank4[0]
    print rank
    
    
    #print 'P',rank[1],data.Point[rank[1]][choice]
    #print 'R',RESULT[rank[1]][choice]
   
        
    #for in2 in range(1,len(rank)):
        #if rank[in2] != rank[0]:#
            ##print 'R-',RESULT[selected_cluster.index(rank[in2])][choice]
            #RESULT[selected_cluster.index(rank[in2])][choice] = 0.0


    for ri in range(1,len(rank)):
        
        re = []
        for ind_p,p in enumerate(RESULT[rank[ri]]):
            if ind_p == choice:
                re.append(0.0)
                continue
            re.append(p)
            
        RESULT[rank[ri]] = re
        
    #print 'P-',rank[1],data.Point[rank[1]][choice]
    #print 'R-',RESULT[rank[1]][choice]
    
    return RESULT

#CHECK-----------------------------------------------------------------------------------
def CHECK(RE):
    
    print 'CHECK--------------'
    E = [0 for i in range(len(RE[0]))]
    for index,r in enumerate(RE):
        E = [sum(x) for x in zip(E, r)]
    if 0 in E:
        print 'RE有0'
    d1 = []
    d0 = []
    for d in range(len(E)):
        if E[d] >1:
            d1.append(d)
        if E[d] == 0:
            d0.append(d)

    print 'len->1(must=0)',len(d1)     

    print 'len-=0',len(d0)
    #print E
    return 0
#convert-----------------------------------------------------------------------------------
def convert(RESULT):
    RE = []
    for ind,res in enumerate(RESULT):
        result = []
        for r in res:
            if r > 0:
                result.append(1)
                continue
            result.append(0)
        RE.append(result)
    #print "RE",RE
    return RE
############################################
for q in range(10):
    print "OK"
```
