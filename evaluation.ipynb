Ensemble Result

```python
import numpy as np
import pandas as pd
import random
from sklearn import metrics
from bokeh.plotting import figure, show, output_file

#ensemble-result=========================================================================
file = 'Draw.txt'

result=[]
with open(file,'r') as fc:
  for line in fc:
    li = line.strip()
    result.append(li.replace(" ", ""))

C = []
for r in result:
  temp = []
  for u in range(len(r)):
    if u%2==1 :
      temp.append(int(r[u]))
  C.append(temp)  

Ensemble = []
for index,i in enumerate(C):
    for j in i:
        if j == 1:
            Ensemble.append(index)

print 'Ensemble\n'
print Ensemble

fc.close()
```
